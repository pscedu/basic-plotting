

# Subplots and Figure Layouts

This notebook demonstrates how to create multiple subplots and arrange figures using `matplotlib`.

This will showcase 

- Basic subplots
- Custom layouts
- Sharing axes
- Saving subplots to disk

```python
import matplotlib.pyplot as plt #basic-plotting
import numpy as np #numerical computing
```

## Example 1: Basic 2x1 Subplots
Two plots stacked vertically.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x) # sine
y2 = np.cos(x) # cosine

# need to create an empty figure to place the plots
fig, axs = plt.subplots(2, 1, figsize=(8,8))  # 2 rows, 1 column

# create 1st subplot
axs[0].plot(x, y1, color='blue', label='sin(x)')
axs[0].set_title('Sine Wave')
axs[0].grid(True)
axs[0].legend()

# create 2nd subplot
axs[1].plot(x, y2, color='red', linestyle='--', label='cos(x)')
axs[1].set_title('Cosine Wave')
axs[1].grid(True)
axs[1].legend()

# save and display plot in notebook
plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x) # sine
y2 = np.cos(x) # cosine

# need to create an empty figure to place the plots
fig, axs = plt.subplots(1, 2, figsize=(8,4))  # 1 row, 1 columns

# create 1st subplot
axs[0].plot(x, y1, color='blue', label='sin(x)')
axs[0].set_title('Sine Wave')
axs[0].grid(True)
axs[0].legend()

# create 2nd subplot
axs[1].plot(x, y2, color='red', linestyle='--', label='cos(x)')
axs[1].set_title('Cosine Wave')
axs[1].grid(True)
axs[1].legend()

# display plot in notebook
plt.tight_layout()
plt.show()
```

## Example 2: 2x2 Subplots with Shared Axes
Four subplots in a grid, sharing x-axis.

```python
fig, axs = plt.subplots(2, 2, figsize=(10,8), sharex=True)

y_values = [np.sin(x), np.cos(x), np.sin(2*x), np.cos(2*x)]
titles = ['sin(x)', 'cos(x)', 'sin(2x)', 'cos(2x)']
colors = ['blue', 'red', 'green', 'purple']

for ax, y, title, color in zip(axs.flatten(), y_values, titles, colors):
    ax.plot(x, y, color=color, linewidth=2)
    ax.set_title(title)
    ax.grid(True)

plt.tight_layout()
plt.show()
```

## Example 3: Custom Figure Layouts with `GridSpec`
Create an asymmetric layout using `GridSpec`.

```python
from matplotlib.gridspec import GridSpec

fig = plt.figure(figsize=(10,6))
gs = GridSpec(3, 3, figure=fig)

# Large subplot spanning first row
ax1 = fig.add_subplot(gs[0, :])
ax1.plot(x, np.sin(x), color='blue')
ax1.set_title('Top Wide Plot')
ax1.grid(True)

# Bottom left subplot
ax2 = fig.add_subplot(gs[1:, 0:2])
ax2.plot(x, np.cos(x), color='red')
ax2.set_title('Bottom Left')
ax2.grid(True)

# Bottom right subplot
ax3 = fig.add_subplot(gs[1:, 2])
ax3.plot(x, np.sin(2*x), color='green')
ax3.set_title('Bottom Right')
ax3.grid(True)

plt.tight_layout()
plt.show()
```

---
### Notes
- Use `plt.subplots(nrows, ncols)` for simple grids.
- Use `GridSpec` for custom layouts and asymmetric arrangements.
- `tight_layout()` avoids overlapping labels.
- `plt.savefig()` saves each figure to disk while `plt.show()` displays it.
