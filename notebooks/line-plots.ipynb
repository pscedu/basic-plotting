

# Line Plots Examples

This notebook demonstrates how to create line plots in Python using `matplotlib`.
We will show multiple examples including single line plots, multiple lines, and plots from a real public dataset.

**Dataset:** We'll use a sample dataset from Seaborn (`tips`) which can be downloaded automatically.

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

## Example 1: Basic Line Plot
Plotting a simple sine wave.

```python
# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create plot
plt.figure(figsize=(8,5))
plt.plot(x, y, color='blue', linewidth=2, label='sin(x)')
plt.title('Basic Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.legend()

# Save the plot to disk
plt.savefig("example1.png", dpi=300, bbox_inches='tight')

# Display the plot
plt.show()
```

```python
# Generate data
x = np.linspace(0, 10, 100)
y = np.tan(x)

# Create plot
plt.figure(figsize=(8,5))
plt.plot(x, y, color='blue', linewidth=2, label='tan(x)')
plt.title('Basic Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.legend()

# Save the plot to disk
plt.savefig("example1.png", dpi=300, bbox_inches='tight')

# Display the plot
plt.show()
```

## Example 2: Multiple Lines
Plot sine and cosine on the same axes.

```python
# Generate data
y1 = np.sin(x)
y2 = np.cos(x)

# Plot
plt.figure(figsize=(8,5))
plt.plot(x, y1, label='sin(x)', color='blue', linewidth=2)
plt.plot(x, y2, label='cos(x)', color='red', linestyle='--', linewidth=2)
plt.title('Multiple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True, linestyle=':')
plt.legend()
plt.show()
```

## Example 3: Line Plot from Public Dataset
We'll use the `tips` dataset from Seaborn, which contains information about restaurant tips. We'll plot `total_bill` over `size` (number of people).

```python
# Load dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Aggregate data: average total_bill per table size
avg_bill = tips.groupby('size')['total_bill'].mean().reset_index()

# Plot
plt.figure(figsize=(8,5))
plt.plot(avg_bill['size'], avg_bill['total_bill'], marker='o', color='green')
plt.title('Average Total Bill vs Table Size')
plt.xlabel('Table Size (number of people)')
plt.ylabel('Average Total Bill ($)')
plt.grid(True)
plt.show()
```

## Example 4: Multiple Lines with Dataset
Compare average `total_bill` by `sex` of the customer.

```python
# Aggregate by sex and size
avg_bill_sex = tips.groupby(['size','sex'])['total_bill'].mean().reset_index()

# Plot
plt.figure(figsize=(8,5))
for sex in avg_bill_sex['sex'].unique():
    data = avg_bill_sex[avg_bill_sex['sex'] == sex]
    plt.plot(data['size'], data['total_bill'], marker='o', label=sex)

plt.title('Average Total Bill by Table Size and Sex')
plt.xlabel('Table Size')
plt.ylabel('Average Total Bill ($)')
plt.legend()
plt.grid(True)
plt.show()
```
